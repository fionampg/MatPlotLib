```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.merge(ride_data, city_data, on="city", how="left")

# Display the data table for preview
pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
## Obtain the x and y coordinates for each of the three city types
#Dataframe for each type of cities
type_urban = pyber_df[pyber_df["type"] == "Urban"].groupby([pyber_df["city"]])
type_suburban = pyber_df[pyber_df["type"] == "Suburban"].groupby([pyber_df["city"]])
type_rural = pyber_df[pyber_df["type"] == "Rural"].groupby([pyber_df["city"]])

#Get x,y,z coordinates, where x = number of rides, y = average of fare, z = number of drivers by city type

# Urban 
urban_x = type_urban["ride_id"].count()
urban_y = type_urban["fare"].mean()
urban_z = type_urban["driver_count"].mean()

# Suburban
suburban_x = type_suburban["ride_id"].count()
suburban_y = type_suburban["fare"].mean()
suburban_z = type_suburban["driver_count"].mean()

# Rural
rural_x = type_rural["ride_id"].count()
rural_y = type_rural["fare"].mean()
rural_z = type_rural["driver_count"].mean()

## Build the scatter plots for each city types
plt.scatter(urban_x, urban_y, label = "Urban", s=urban_z * 10, color=["coral"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(suburban_x, suburban_y, label = "Suburban", s=suburban_z  * 10, color=["lightskyblue"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(rural_x, rural_y, label = "Rural", s=rural_z * 10, color=["gold"], edgecolor="black", alpha = 0.75, marker="o")
plt.grid()

## Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

## Create a legend
legend = plt.legend(fontsize = 8, title= "City Types", loc="best") 

# Adjust the legend 
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

## Incorporate a text label regarding circle size
plt.text(0,0,"Note: \nCircle size correlates with driver count per city.", fontsize = 10)

# Save Figure
plt.savefig("../Images/PyberRide.png", bbox_inches="tight")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
citytype_group = pyber_df.groupby(["type"])
sumfare = citytype_group["fare"].sum()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0,0,0.1)
colors = ["gold","lightskyblue","lightcoral"]

plt.pie(sumfare, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("../Images/TotalFaresbyCityType.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
totalrides_count = citytype_group["ride_id"].count()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(totalrides_count, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("../Images/TotalRidesbyCityType.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents, get data only from city data.csv
totaldrivers_city = city_data.groupby(['type'])
totaldrivers = totaldrivers_city['driver_count'].sum()

# Build Pie Charts
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(totaldrivers, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("../Images/TotalDriversbyCityType.png")
```

```python
# Show Figure
plt.show()
```

##**Observational Trends**

Pyber is currently present in urban, suburban and rural areas. Because of the different types of areas, average fare price, number of rides, and number of drivers for each type are different by type. 

**1.** In urban areas, there are more rides, about 68% of Pyber's total rides, and more drivers, about 81% of Pyber's total drivers. Rural areas account for about 5% and 3% in the same categories. 

**2.**  From the data, we can also see that total fares in urban areas account for 63% of total fares. On the other hand, suburban and rural areas account for 30% and 7%, respectively.

**3.** From these observations we can conclude that there is an inverse relationship between average fare and number of trips in rural and urban areas: in urban areas, the average fare is lower, with more drivers take more trips in urban areas. In rural areas, the average fare is higher, and there are fewer drivers who in turn take fewer trips. 

This could be because there are more Pyber drivers in urban areas than in rural areas, which means there is more competition between drivers. And, competition pushes average fares down. 

Another reason for lower fares in urban areas is passengers take shorter trips (distance-wise) more frequently than those in rural areas. However, this assumption is not confirmed since we do not have data on the duration of the trip.
